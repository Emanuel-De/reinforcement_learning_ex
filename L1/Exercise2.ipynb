```
def ac_va_iter(pi, Q, r, p, gamma, thresh):
  delta = 1
  #1=up, 2=down, 3=left, 4=right
  while delta > thresh:
      Q_0 = Q.copy()
      for s in range(Q.shape[0]):
        if s != 5:
          for a in range(Q.shape[1]):
            p_vec = p[a][s]
            #Q[s,a] = r[s]-int(s!=3)*50 + gamma * np.sum(p_vec[p_vec>0]*np.max(Q[p_vec>0], axis=1))      #tryout for adding a negativ reward component
            Q[s,a] = r[s] + gamma * np.sum(p_vec[p_vec>0]*np.max(Q[p_vec>0], axis=1))
      delta = np.max(np.abs(Q - Q_0))

  pi = np.argmax(Q, axis=1).reshape(3,4) + 1
  pi[1,1] = 0

  return pi, Q
```
