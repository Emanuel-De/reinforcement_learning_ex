```
def simulator(s, a):

  r = np.array([0, 0, 0, +1,
              0, 0, 0, -100,
              0, 0, 0, 0])
  
  p = movement()

 #0=up, 1=down, 2=left, 3=right

  snext = np.random.choice(12 ,p=p[a][s])
  
  return snext, r[snext]
```

```
def q_learning(Q, s, gamma):
  itrations = 2000
  #0=up, 1=down, 2=left, 3=right
  Q_0 = Q.copy()

  for i in range(itrations):

    epsilon = 1/(0.0002 * i + 1)
    alpha = 1/(0.0005 * i + 1000)

    if np.random.rand() < epsilon:
      a= np.random.randint(4)
    else:
      a = np.argmax(Q[s])

    snext, r = simulator(s, a)
    Q[s,a] = Q[s,a] + alpha*(r + gamma * np.max(Q[snext]) - Q[s,a])
    s = snext

  return Q, i
```

```
# declaring variables
gamma = 0.9 #discount factor for future rewards

s_init = np.random.randint(low=0, high=12, size=1).astype(np.int)
if s_init == 5: s_init = s_init + np.random.randint(-5, 7)
s = s_init[0]

Q = np.zeros((12, 4))

Q, i = q_learning(Q, s, gamma)


print(f'Number of iterations: {i+1}')

pi = np.argmax(Q, axis=1).reshape(3,4) + 1
pi[1,1] = 0

pi_print = pi.reshape(3,4)

Q_max = np.amax(Q[:], axis=1)
print(f'Q_max: \n{Q_max.reshape(3,4)}\n\n')

Q_print_up = Q[:,0].reshape(3,4)
Q_print_down = Q[:,1].reshape(3,4)
Q_print_left = Q[:,2].reshape(3,4)
Q_print_right = Q[:,3].reshape(3,4)

print(f'Q Table up: \n{Q_print_up}')
print(f'Q Table down: \n{Q_print_down}')
print(f'Q Table left: \n{Q_print_left}')
print(f'Q Table right: \n{Q_print_right}')

print(f'\nFinal Policy: \n{pi_print}\n0=Nothing, 1=up, 2=down, 3=left, 4=right')

pi_print_arrows = pi.reshape(3,4).astype(str)
pi_print_arrows[pi_print_arrows=='1']='\u2191'
pi_print_arrows[pi_print_arrows=='2']='\u2193'
pi_print_arrows[pi_print_arrows=='3']='\u2190'
pi_print_arrows[pi_print_arrows=='4']='\u2192'
print(f'\nFinal Policy: \n{pi_print_arrows}')
```
